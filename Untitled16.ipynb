```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
n = np.arange(1, 10000)
probability = 1-(1-(1/n))**n
fig, ax = plt.subplots(figsize = (15,10))
ax.plot(n, probability)
ax.set(xscale = "log", xlabel = "observations", ylabel = "Probability")
```
